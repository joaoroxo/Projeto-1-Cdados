```python
transpor_m = dadospib_mercosul.transpose()
transpor_qnt_m = transpor_m[:-2]
mercosul1 = transpor_qnt_m.loc[(transpor_qnt_m.index>1990) & (transpor_qnt_m.index<2017),:].transpose()
soma_m = mercosul1.sum()
```

```python
def reta(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear entre X3 e X4:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
    plt.scatter(dados_x, dados_y, c=c, alpha=0.8)
    plt.title('x versus y')
    plt.xlabel('eixo x')
    plt.ylabel('eixo y')

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()
```

```python
plt.scatter(dadosexp_country.loc[dadosexp_country.MERCOSUL=="Mercosul",2010], dadospib_country.loc[dadospib_country.MERCOSUL=="Mercosul", 2010], color="r")
reta(dadosexp_country.loc[dadosexp_country.MERCOSUL=="Mercosul",2010], dadospib_country.loc[dadospib_country.MERCOSUL=="Mercosul",2010], "r")
plt.scatter(dadosexp_country.loc[dadosexp_country.MERCOSUL=="Associados",2010], dadospib_country.loc[dadospib_country.MERCOSUL=="Associados", 2010], color="b")
reta(dadosexp_country.loc[dadosexp_country.MERCOSUL=="Associados",2010], dadospib_country.loc[dadospib_country.MERCOSUL=="Associados",2010], "b")
```
